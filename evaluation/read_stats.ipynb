```python
rtg = [0.9, 1.0]
rtg2 = [0, 0.5, 1.5, 2.0]
mean = {}
y = {}
x = np.arange(1,26)

for r in rtg:
    mean[str(r)] = stats[stats['target_return']==r]['total_return'].iloc[-25:].mean()
    y[str(r)] = stats[stats['target_return']==r]['total_return'].iloc[-25:].values
for r in rtg2:
    mean[str(r)] = stats2[stats2['target_return']==r]['total_return'].iloc[-25:].mean()
    y[str(r)] = stats2[stats2['target_return']==r]['total_return'].iloc[-25:].values

print(y.keys())
print(mean)

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True)
fig.suptitle('ORCA Returns')
axs[0,0].plot(x, y['0'])
axs[0,0].axhline(y=mean['0'], color='red', linestyle='--', label='mean')
axs[0,0].set_title('Target Return: 0.0')
axs[0,1].plot(x, y['0.5'])
axs[0,1].axhline(y=mean['0.5'], color='red', linestyle='--', label='mean')
axs[0,1].set_title('Target Return: 0.5')
axs[0,2].plot(x, y['0.9'])
axs[0,2].axhline(y=mean['0.9'], color='red', linestyle='--', label='mean')
axs[0,2].set_title('Target Return: 0.9')

axs[1,0].plot(x, y['1.0'])
axs[1,0].axhline(y=mean['1.0'], color='red', linestyle='--', label='mean')
axs[1,0].set_title('Target Return: 1.0')
axs[1,1].plot(x, y['1.5'])
axs[1,1].axhline(y=mean['1.5'], color='red', linestyle='--', label='mean')
axs[1,1].set_title('Target Return: 1.5')
axs[1,2].plot(x, y['2.0'])
axs[1,2].axhline(y=mean['2.0'], color='red', linestyle='--', label='mean')
axs[1,2].set_title('Target Return: 2.0')
fig.show()
```

```python
rtg = [0.9, 1.0]
mean_k20, mean_k5 = {}, {}
y_k20, y_k5 = {}, {}

for r in rtg:
    mean_k20[str(r)] = stats[stats['target_return']==r]['total_return'].mean()
    y_k20[str(r)] = stats[stats['target_return']==r]['total_return'].values
    mean_k5[str(r)] = stats_k5[stats_k5['target_return']==r]['total_return'].mean()
    y_k5[str(r)] = stats_k5[stats_k5['target_return']==r]['total_return'].values


print(y_k20.keys())
print(mean_k20, mean_k5)

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
fig.suptitle('ORCA Returns with Different Context')
axs[0,0].plot(y_k20['0.9'])
axs[0,0].axhline(y=mean_k20['0.9'], color='red', linestyle='--', label='mean')
axs[0,0].set_title('Target Return: 0.9 | k = 20')
axs[0,1].plot(y_k5['0.9'])
axs[0,1].axhline(y=mean_k5['0.9'], color='red', linestyle='--', label='mean')
axs[0,1].set_title('Target Return: 0.9 | k = 5')

axs[1,0].plot(y_k20['1.0'])
axs[1,0].axhline(y=mean_k20['1.0'], color='red', linestyle='--', label='mean')
axs[1,0].set_title('Target Return: 1.0 | k = 20')
axs[1,1].plot(y_k5['1.0'])
axs[1,1].axhline(y=mean_k5['1.0'], color='red', linestyle='--', label='mean')
axs[1,1].set_title('Target Return: 1.0 | k = 5')
```

```python
rtg3 = [-2.0, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0]
mean, std, = [], []

for r in rtg3:
    mean.append(stats2[stats2['target_return']==r]['total_return'].iloc[:100].mean())
    std.append(np.std(stats2[stats2['target_return']==r]['total_return'].iloc[0:100].values))

plt.errorbar(rtg3, mean, yerr=std, fmt='-o',capsize=4)
plt.title('ORCA 500 Episode Average Returns')
plt.xlabel('Target Return')
plt.ylabel('Average Total Return')
plt.show()
```
